```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def mae(y, yhat):
    error = np.abs(yhat - y)
    return np.mean(error)
    
```

```python
def mse(y, y_hat):
    error = np.abs(y_hat - y)
    return np.mean(error ** 2)
```

```python
def rmse(y, y_hat):
    return np.sqrt(mse(y, y_hat))
```

```python
def r2(y, y_hat):
    ss_res = np.sum((y - y_hat) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - ss_res/ss_tot
```

```python
rng = np.random.default_rng()
y = rng.normal(size=100)
y_hat = y + rng.normal(scale=0.5,size=100)
```

```python
for name, mine, ref in [
    ("MAE", mae(y, y_hat), mean_absolute_error(y, y_hat)),
    ("MSE", mse(y, y_hat), mean_squared_error(y, y_hat)),
    ("RMSE", rmse(y, y_hat), np.sqrt(mean_squared_error(y, y_hat))),
    ("R²",   r2(y, y_hat),   r2_score(y, y_hat)),
]:
    ok = np.isclose(mine, ref)
    print(f"{name:5} mine={mine:.6f}  ref={ref:.6f}  {'✓' if ok else '✗'}")
```
